```python
import numpy as np
import matplotlib.pyplot as plt

# x, y  
x = np.linspace(0, 100, 101)
y = np.linspace(0, 100, 101)
X, Y = np.meshgrid(x, y)

#     (0 10  )
Z = np.random.rand(101, 101) * 10

#   
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, cmap='viridis')  #   
plt.colorbar(label='Height')  #  
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Height Map Overlay')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#  
x = np.linspace(-5, 5, 20)  # x 
y = np.linspace(-5, 5, 20)  # y 
X, Y = np.meshgrid(x, y)  #  

# u, v  
u = np.sin(X)  # X   u   ( )
v = np.cos(Y)  # Y   v   ( )

# 
plt.figure(figsize=(8, 6))
plt.quiver(X, Y, u, v, scale=20)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Vector Field Overlay')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Z   ( )
Z = np.random.rand(101, 10) * 10

# Z   u, v  
u = np.gradient(Z, axis=0)  # Z   x  u  
v = np.gradient(Z, axis=1)  # Z   y  v  

#  
x = np.linspace(0, 1, Z.shape[1])
y = np.linspace(0, 1, Z.shape[0])
X, Y = np.meshgrid(x, y)

#  
plt.figure(figsize=(8, 6))
plt.quiver(X, Y, u, v)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Vector Field Overlay')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd

#   
width = 25.950
height = 32.950

#   
grid_size = 5

#      
def generate_random_coords_in_grid(x_start, y_start, x_end, y_end, num_points):
    random_x = np.random.randint(x_start, x_end, num_points)
    random_y = np.random.randint(y_start, y_end, num_points)
    return random_x, random_y

#    
num_points_per_grid = 1
random_x_all = []
random_y_all = []
for x_start in range(-int(width/2), int(width/2), grid_size):
    for y_start in range(-int(height/2), int(height/2), grid_size):
        x_end = min(x_start + grid_size, width)
        y_end = min(y_start + grid_size, height)
        num_points = min(num_points_per_grid, (x_end - x_start) * (y_end - y_start))
        random_x, random_y = generate_random_coords_in_grid(x_start, y_start, x_end, y_end, num_points)
        random_x_all.extend(random_x)
        random_y_all.extend(random_y)

#   
for i, (x, y) in enumerate(zip(random_x_all, random_y_all), 1):
    print(f"Point {i}: ({x}, {y})")

test_pos = pd.DataFrame([random_x_all,random_y_all]).transpose()
```

```python
import pandas as pd
import numpy as np

#  
repeat_count = 30

# x, y  
x_range = np.arange(-5, 6)
y_range = np.arange(-5, 6)

#   
df = pd.DataFrame(columns=['Row', 'Col','test_id'])

#   30    
for x in x_range:
    for y in y_range:
        #   30  
        coordinates = [(x, y, n) for n in range(len(test_pos))]
        #  
        df = pd.concat([df,pd.DataFrame(coordinates, columns=['Row', 'Col','test_id'])], ignore_index=True)

#  
df
shot_x_size = 25.950
shot_y_size = 32.950
shot_x_offset = shot_x_size*0
shot_y_offset = shot_y_size*0.5
df['shot_X'] = df['Row']*shot_x_size + shot_x_offset
df['shot_Y'] = df['Col']*shot_y_size + shot_y_offset
df['x_cor'] = ''
df['y_cor'] = ''
for i in range(len(df)):
    df['x_cor'].iloc[i] = test_pos.iloc[df['test_id'].iloc[i],0]
    df['y_cor'].iloc[i] = test_pos.iloc[df['test_id'].iloc[i],1]
```

```python
df['wafer_x'] = df['shot_X']+df['x_cor']
df['wafer_y'] = df['shot_Y']+df['y_cor']
df['radius'] =(df['wafer_x']**2)+(df['wafer_y']**2)
df['radius'] = df['radius'].apply(lambda x:x**0.5)
df = df[df['radius']<150]
df
```

```python
import numpy as np
import matplotlib.pyplot as plt

# x, y 좌표 생성
x = np.linspace(0, 100, 101)
y = np.linspace(0, 100, 101)
X, Y = np.meshgrid(x, y)

# 임의의 높이 데이터 생성 (0부터 10까지의 무작위 값)
Z = np.random.rand(101, 101) * 10

# 높이 맵 시각화
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, cmap='viridis')  # 등고선 맵으로 표시
plt.colorbar(label='Height')  # 컬러바 추가
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Height Map Overlay')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 생성
x = np.linspace(-5, 5, 20)  # x 좌표
y = np.linspace(-5, 5, 20)  # y 좌표
X, Y = np.meshgrid(x, y)  # 격자 생성

# u, v 벡터 생성
u = np.sin(X)  # X 좌표에 대한 u 벡터 값 (임의의 예시)
v = np.cos(Y)  # Y 좌표에 대한 v 벡터 값 (임의의 예시)

# 그리기
plt.figure(figsize=(8, 6))
plt.quiver(X, Y, u, v, scale=20)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Vector Field Overlay')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Z 값을 생성합니다 (임의의 예시)
Z = np.random.rand(101, 10) * 10

# Z 값을 사용하여 u, v 벡터를 생성합니다
u = np.gradient(Z, axis=0)  # Z 값을 사용하여 x 방향의 u 벡터를 생성합니다
v = np.gradient(Z, axis=1)  # Z 값을 사용하여 y 방향의 v 벡터를 생성합니다

# 격자 생성
x = np.linspace(0, 1, Z.shape[1])
y = np.linspace(0, 1, Z.shape[0])
X, Y = np.meshgrid(x, y)

# 벡터 플롯
plt.figure(figsize=(8, 6))
plt.quiver(X, Y, u, v)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Vector Field Overlay')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd

# 영역 크기 설정
width = 25.950
height = 32.950

# 격자의 크기 설정
grid_size = 5

# 격자 내 랜덤 좌표 생성 함수
def generate_random_coords_in_grid(x_start, y_start, x_end, y_end, num_points):
    random_x = np.random.randint(x_start, x_end, num_points)
    random_y = np.random.randint(y_start, y_end, num_points)
    return random_x, random_y

# 격자별로 랜덤 좌표 생성
num_points_per_grid = 1
random_x_all = []
random_y_all = []
for x_start in range(-int(width/2), int(width/2), grid_size):
    for y_start in range(-int(height/2), int(height/2), grid_size):
        x_end = min(x_start + grid_size, width)
        y_end = min(y_start + grid_size, height)
        num_points = min(num_points_per_grid, (x_end - x_start) * (y_end - y_start))
        random_x, random_y = generate_random_coords_in_grid(x_start, y_start, x_end, y_end, num_points)
        random_x_all.extend(random_x)
        random_y_all.extend(random_y)

# 생성된 좌표 출력
for i, (x, y) in enumerate(zip(random_x_all, random_y_all), 1):
    print(f"Point {i}: ({x}, {y})")

test_pos = pd.DataFrame([random_x_all,random_y_all]).transpose()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# 주어진 데이터
X_train = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)  # 3월부터 8월까지의 월 수
y_train = np.array([0.02, 0.02, 0.02, 0.02, 0.02, 0.05])  # 해당 월의 값

# # 다항 회귀 모델 생성 및 학습
# poly = PolynomialFeatures(degree=3)  # 3차 다항식 사용
# X_poly = poly.fit_transform(X_train)
# poly.fit(X_poly, y_train)
# model = LinearRegression()
# model.fit(X_poly, y_train)

# 다항 회귀 모델 생성 및 학습
poly = PolynomialFeatures(degree=3)  # 3차 다항식 사용
X_train_poly = poly.fit_transform(X_train)
model = LinearRegression()
model.fit(X_train_poly, y_train)

# 9월의 값을 예측
X_test = np.array([[7]])  # 9월의 월 수
X_test_poly = poly.transform(X_test)
predicted_value = model.predict(X_test_poly)

# 그래프 그리기
plt.scatter(X_train, y_train, color='blue', label='Train Data')  # 월별 데이터
plt.plot(X_train, model.predict(poly.fit_transform(X_train)), color='red', label='Regression Curve')  # 회귀선
plt.scatter(X_test, predicted_value, color='green', label='Predicted Value for September')  # 9월 값 예측
plt.xlabel('Month')
plt.ylabel('Value')
plt.title('Polynomial Regression for Monthly Data')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# 주어진 데이터
X_train = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)  # 3월부터 8월까지의 월 수
y_train = np.array([0.02, 0.02, 0.02, 0.02, 0.02, 0.05])  # 해당 월의 값

# 지수 함수 적용
X_train_exp = np.exp(X_train)
y_train_exp = np.exp(y_train)

# 선형 회귀 모델 생성 및 학습
model = LinearRegression()
model.fit(X_train_exp, y_train_exp)

# 9월의 값을 예측
X_test = np.array([[7]])  # 9월의 월 수
X_test_exp = np.exp(X_test)
predicted_value_exp = model.predict(X_test_exp)

# 그래프 그리기
plt.scatter(X_train, y_train, color='blue', label='Train Data')  # 월별 데이터
plt.plot(X_train, model.predict(X_train_exp), color='red', label='Regression Curve')  # 회귀선
plt.scatter(X_test, predicted_value_exp, color='green', label='Predicted Value for September')  # 9월 값 예측
plt.xlabel('Month')
plt.ylabel('Value')
plt.title('Exponential Regression for Monthly Data')
plt.legend()
plt.show()
```

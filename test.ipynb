```python
import numpy as np
import matplotlib.pyplot as plt

# 예제 데이터 생성
np.random.seed(0)
X = np.random.rand(700)  # X 좌표
Y = np.random.rand(700)  # Y 좌표
U = 0.02 * X  # X 방향 벡터
V = 0.03 * Y  # Y 방향 벡터

# Quiver plot 그리기
plt.figure()
plt.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1, color='b', label='Vectors')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')

# 벡터 그래프 추가 (설명용)
plt.scatter(X, Y, color='r', label='Points')
plt.legend()

plt.xlim(0, 1)
plt.ylim(0, 1)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 원의 중심 좌표
center_x = 0
center_y = 0

# 원의 반지름 (지름의 절반)
radius = 150  # mm

# 벡터의 크기 (방향은 중심에서 원의 테두리 방향으로)
vector_magnitude = 50

# 벡터의 방향 (예시: 오른쪽 위 방향)
vector_direction_x = 1
vector_direction_y = 1

# 원 위의 점들을 생성
theta = np.linspace(0, 2 * np.pi, 100)
x = center_x + radius * np.cos(theta)
y = center_y + radius * np.sin(theta)

# 벡터의 끝점 좌표 계산
vector_end_x = center_x + vector_magnitude * vector_direction_x
vector_end_y = center_y + vector_magnitude * vector_direction_y

# Quiver plot 그리기
plt.figure(figsize=(8, 8))
plt.scatter(x, y, color='b', label='Circle')
plt.quiver(center_x, center_y, vector_direction_x, vector_direction_y, angles='xy', scale_units='xy', scale=1, color='r', label='Vector')
plt.xlim(-radius - 50, radius + 50)
plt.ylim(-radius - 50, radius + 50)
plt.grid()
plt.show()
```
